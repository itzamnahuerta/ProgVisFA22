```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../hw-assignment/data/events-US-1980-2022.csv')
```

```python
df.head(2)
```

```python
# 12 features and 338 records in dataset
df.shape
```

```python
df.head(3)
```

```python
#  replacing empty spaces in column names with underscore and lowercasing names
df.columns = df.columns.str.replace(' ','_')
df.columns= df.columns.str.lower()
```

```python
#  renaming column name
df.rename(columns={'total_cpi-adjusted_cost_(millions_of_dollars)': "total_cpi_adjusted_cost_by_millions"}, inplace=True)
```

```python
# creating new columns from the begin and end date of natural disaster
df['begin_year'] = df['begin_date'].astype(str).str[:4]
df['begin_month'] = df['begin_date'].astype(str).str[4:6]
df['begin_day']  = df['begin_date'].astype(str).str[6:8]

df['end_year'] = df['end_date'].astype(str).str[:4]
df['end_month'] = df['end_date'].astype(str).str[4:6]
df['end_day']  = df['end_date'].astype(str).str[6:8]
```

```python
# dataframe with updated columns
df.head()
```

```python
# looking at data types
df.info()
```

```python
df.head()
```

```python
# removing records where total cpi adjusted cost is TBD and updating object data type to float
df = df[df['total_cpi_adjusted_cost_by_millions'] != 'TBD']

df['total_cpi_adjusted_cost_by_millions'] = df['total_cpi_adjusted_cost_by_millions'].astype(float)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
# verifying update from above
# df.info()
```

```python
df.drop('begin_date', inplace=True, axis=1)
```

```python
df.drop('end_date', inplace=True, axis=1)
```

```python
# updating type from string to integer

df['begin_date'] = df['begin_month'].astype(str) + '-' + df['begin_day'].astype(str) + '-' + df['begin_year'].astype(str)

df['begin_date'] = pd.to_datetime(df.begin_date, format='%m-%d-%Y')
```

```python
df['end_date'] = df['end_month'].astype(str) + '-' + df['end_day'].astype(str) + '-' + df['end_year'].astype(str)
```

```python
df['end_date'] = pd.to_datetime(df.end_date,format='%m-%d-%Y')
```

```python
df['name'].value_counts()
```

```python
df['name'] = df['name'].str.split('(',1).str[0]
```

```python
df['total_cpi_adjusted_cost_by_millions'] = df['total_cpi_adjusted_cost_by_millions'].apply(np.round)
```

```python
df['total_cpi_adjusted_cost_by_millions'] = df['total_cpi_adjusted_cost_by_millions'].astype(int)
```
